```python
import numpy as np
```

```python
A=np.array([[2,4],[1,3],[0,0],[0,0]])
```

```python
dot=np.dot(A,A.T)
dot
```

```python
eigval,eigvec=np.linalg.eig(dot)
```

```python
eigval
```

```python
eigvec
```

```python
dot2=np.dot(A.T,A)
```

```python
eigval1,eigvec1=np.linalg.eig(dot2)
```

```python
eigval1
```
