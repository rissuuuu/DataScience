```python
calcofi=bottle.loc[:,'Depthm':'O2Satq']
```

```python
calcofi.isnull().sum()
```

```python
fig,ax=pl.subplots(1,2)
sns.distplot(calcofi.T_degC.dropna(),ax=ax[0])
ax[0].set_title('Dropna')
sns.distplot(calcofi.T_degC.fillna(calcofi.T_degC.mean()),ax=ax[1])
ax[1].set_title("Fillna with mean")
```

```python
fig, ax=plt.subplots(1,2)
sns.distplot(calcofi.O2Satq.dropna(), ax=ax[0],kde=False)
ax[0].set_title("Dropna")
sns.distplot(calcofi.O2Satq.fillna(calcofi.O2Satq.mean()), ax=ax[1],kde=False)
ax[1].set_title("Fillna mean")
```
